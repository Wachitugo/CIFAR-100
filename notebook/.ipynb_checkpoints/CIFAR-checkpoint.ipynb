```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
data = unpickle('../data/test')  # Ajusta la ruta según tu estructura
meta_data = unpickle('../data/meta')  # Archivo meta para los nombres de las categorías
```

```python
images = data[b'data'].reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)
fine_labels = data[b'fine_labels']
fine_label_names = [name.decode('utf-8') for name in meta_data[b'fine_label_names']]
```

```python
unique_labels = np.unique(fine_labels)
num_categories = len(unique_labels)

# Configurar la visualización
plt.figure(figsize=(20, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for i, label in enumerate(unique_labels):
    # Encontrar el primer índice de esta categoría
    idx = np.where(fine_labels == label)[0][0]
    
    # Mostrar la imagen
    plt.subplot(10, 10, i + 1)  # Ajusta el grid según el número de categorías
    plt.imshow(images[idx])
    plt.title(f"{label}: {fine_label_names[label]}", fontsize=8)
    plt.axis('off')

plt.show()
```
