# CIFAR-100

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
data_train = unpickle('../data/train')  
meta_data = unpickle('../data/meta')  
data_test = unpickle('../data/test')  
```

```python
images_train = data_train[b'data'].reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)  
images_test = data_test[b'data'].reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)

fine_labels_train = data_train[b'fine_labels']
fine_labels_test = data_test[b'fine_labels']
fine_label_names = [name.decode('utf-8') for name in meta_data[b'fine_label_names']]
```

```python
unique_labels = np.unique(fine_labels_train)
num_categories = len(unique_labels)

plt.figure(figsize=(20, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for i, label in enumerate(unique_labels):
    idx = np.where(fine_labels_train == label)[0][0]
    
    plt.subplot(10, 10, i + 1)
    plt.imshow(images_train[idx])  
    plt.title(f"{label}: {fine_label_names[label]}", fontsize=8)
    plt.axis('off')

plt.suptitle("una imagen por categoria :v", fontsize=16)
plt.show()
```

```python
print(images_train.shape)
print(images_test.shape)
print(len(fine_labels_train))
print(len(fine_labels_test))
```

```python
from tensorflow.keras.utils import to_categorical

X_train = images_train.reshape(50000,32*32*3)
X_train = X_train / 255
print(X_train.shape)

X_test = images_test.reshape(10000,32*32*3)
X_test = X_test / 255
print(X_test.shape)

y_train = to_categorical(fine_labels_train, 100)
y_test = to_categorical(fine_labels_test, 100)

print(y_train.shape)
print(y_test.shape)
```
